```python
# your imports here
import nltk
from nltk.corpus import wordnet as wn
import pandas as pd
import random
```

### Vanilla N-Gram Language Model - Valid Word Generation

```python
# code here!
vanilla_generation = []

with open('spooky_vanilla_3_char.txt', 'r') as file:
    file_contents = file.read()
    vanilla_generation = file_contents.split()
    
processed_vanilla_generation = [word for word in vanilla_generation if word not in ('.', ',', '!', '?')]

valid_word_count = 0
valid_words = set()
invalid_words = set()

for word in processed_vanilla_generation:
    syns = wn.synsets(word)
    if len(syns) > 0:
        valid_word_count += 1
        valid_words.add(word)
    else:
        invalid_words.add(word)
        
print("Valid Word Generation: " + str(valid_word_count) + "/" + str(len(processed_vanilla_generation)) + " ({:.2f}%)".format(valid_word_count / len(processed_vanilla_generation) * 100))
```

```python
train = pd.read_csv('spooky_author_train.csv')
train_word_vocab = set(train['text'].apply(lambda x: nltk.word_tokenize(x.lower())).sum())
```

```python
new_valid_count = 0
for word in valid_words:
    if word not in train_word_vocab:
        new_valid_count += 1
        
print("New Valid Words: " + str(new_valid_count) + "/" + str(len(processed_vanilla_generation)) + " ({:.2f}%)".format(new_valid_count / len(processed_vanilla_generation) * 100))
```

### Neural - Valid Word Generation

```python
neural_generation = []

with open('char_sents.txt', 'r') as file:
    file_contents = file.read()
    neural_generation = file_contents.split()
    
processed_neural_generation = [word for word in neural_generation if word not in ('.', ',', '!', '?')]

neural_valid_word_count = 0
neural_valid_words = set()
neural_invalid_words = set()

for word in processed_neural_generation:
    syns = wn.synsets(word)
    if len(syns) > 0:
        neural_valid_word_count += 1
        neural_valid_words.add(word)
    else:
        neural_invalid_words.add(word)
        
print("Valid Word Generation: " + str(neural_valid_word_count) + "/" + str(len(processed_neural_generation)) + " ({:.2f}%)".format(neural_valid_word_count / len(processed_neural_generation) * 100))
```

```python
new_neural_valid_count = 0 
for word in neural_valid_words:
    if word not in train_word_vocab:
        new_neural_valid_count += 1
        
print("New Valid Words: " + str(new_neural_valid_count) + "/" + str(len(processed_neural_generation)) + " ({:.2f}%)".format(new_neural_valid_count / len(processed_neural_generation) * 100))
```
